```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
```

```python
#load datasets
path_1 = 'data/depp-rers-2017-donnees-fiche-02-06-810812.xls'
path_2 = 'data/fr-en-baccalaureat-par-academie.csv'
path_3 = 'data/fr-en-college-effectifs-niveau-sexe-lv.csv'
path_4 = 'data/fr-en-dnb-par-etablissement.csv'
path_5 = 'data/fr-en-reussite-au-baccalaureat-origine-sociale.csv'

taille_collège_lycée_data = pd.read_excel(path_1)
bac_academie_data = pd.read_csv(path_2, delimiter=';')
college_effectif_data = pd.read_csv(path_3, delimiter=';')
brevet_établissement_data = pd.read_csv(path_4, delimiter=';')
bac_origine_data = pd.read_csv(path_5, delimiter=';')
```

# Exploration bac / lycée data

## 1 - Bac origine Data

```python
# Basic stats
print(f'Number of rows : {bac_origine_data.shape[0]}\n')

print('Display of dataset: ')
display(bac_origine_data.head())

print('Basics statistics :')
data_desc = bac_origine_data.describe(include='all')
display(data_desc)

print('Percentage of missing values: ')
display(100 * bac_origine_data.isnull().sum() / bac_origine_data.shape[0])

missing_values = bac_origine_data.isnull().any().any()
if missing_values :
    print('There are missing values in the dataset')
else :
    print('There are no missing values in the dataset')
```

```python
column_feature = [
    "Nombre d'admis au baccalaureat général",
    "Nombre d'admis au baccalauréat technologique",
    "Nombre d'admis au baccalauréat professionnel",
    "Nombre d'admis au baccalauréat",
    "Origine sociale"
]

fig = make_subplots(rows=3, cols=2)
row = 1
col = 1
for column in column_feature:           
    fig.add_trace(go.Histogram(x=bac_origine_data[column], name=column),row=row, col=col)
    col += 1 
    if col >2:
        row += 1 
        col =1
fig.update_layout(title= dict(text='Distribution', x=0.5))
fig.update_layout(autosize=False, width=1200, height=1000)

fig.show()
```
