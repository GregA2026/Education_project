```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
```

```python
#load dataset
path = 'reussite_bac_origine_sociale.csv'
df = pd.read_csv(path, delimiter=';')
```

```python
# Basic stats
print(f'Number of rows : {df.shape[0]}\n')

print('Display of dataset: ')
display(df.head())

print('Basics statistics :')
data_desc = df.describe(include='all')
display(data_desc)

print('Percentage of missing values: ')
display(100 * df.isnull().sum() / df.shape[0])

missing_values = df.isnull().any().any()
if missing_values :
    print('There are missing values in the dataset')
else :
    print('There are no missing values in the dataset')
```

```python
column_feature = ["Origine sociale", "Nombre d'admis au baccalaureat général","Nombre d'admis au baccalauréat technologique", 
                  "Nombre d'admis au baccalauréat professionnel", "Nombre d'admis au baccalauréat"                  ]

fig = make_subplots(rows=3, cols=2)
row = 1
col = 1
for column in column_feature:           
    fig.add_trace(go.Histogram(x=df[column], name=column),row=row, col=col)
    col += 1 
    if col >2:
        row += 1 
        col =1
fig.update_layout(title= dict(text='Distribution', x=0.5))
fig.update_layout(autosize=False, width=1200, height=1000)

fig.show()
```
