```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('ETL_Pearson.xlsx')
```

```python
df.head()
```

```python
# Calcul de la corrélation de Pearson
pearson_coef, p_value = stats.pearsonr(df['Taux_Reussite'], df['Taux_mention_TB'])

print(f"Coefficient de Pearson : {pearson_coef:.4f}")
print(f"P-value : {p_value:.4e}")
```

```python
# Création du graphique
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Taux_Reussite', y='Taux_mention_TB', scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Relation entre Taux de Réussite et Taux de Mention TB par Académie')
plt.xlabel('Taux de Réussite')
plt.ylabel('Taux de Mention TB')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
