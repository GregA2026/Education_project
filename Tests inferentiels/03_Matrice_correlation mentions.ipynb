```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Chargement des données
df = pd.read_excel('ETL_mentions.xlsx')
```

```python
df.head()
```

```python
# 2. Création de la matrice de contingence globale
# On croise le Sexe et l'Attribut en sommant la colonne Valeur
matrice_globale = pd.pivot_table(df, 
                                 index='Sexe', 
                                 columns='Mention', 
                                 values='TOTAL', 
                                 aggfunc='sum')
```

```python
# 3. Affichage de la matrice brute
print("--- Matrice de Contingence Globale (Effectifs) ---")
print(matrice_globale)
```

```python
# 4. Visualisation avec une Heatmap (très parlant pour un jury)
plt.figure(figsize=(10, 6))
sns.heatmap(matrice_globale, annot=True, fmt='g', cmap='YlGnBu')
plt.title('Répartition globale des résultats par Sexe')
plt.show()

# optionnel : Exporter la matrice en Excel pour ton rapport
# matrice_globale.to_excel("matrice_contingence_globale.xlsx")
```
