```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Chargement des données
df = pd.read_excel('ETL_reussites.xlsx')
```

```python
df.head()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Utiliser C(Sexe) garantit que Python traite bien Homme/Femme comme des groupes distincts
model = ols('Taux_reussite ~ C(Libellé_académie)', data=df).fit()

# Calcul de la table ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```

```python
import plotly.express as px

# Inversion de x et y pour l'orientation horizontale
fig = px.box(df, 
             x="Taux_reussite",          # Devient l'axe horizontal
             y="Libellé_académie",       # Devient l'axe vertical
             color="Sexe",
             points="all", 
             title="Distribution du Taux de Réussite par Académie et Sexe",
             labels={"Taux_reussite": "Taux de Réussite (%)", "Libellé_académie": "Académie", "Sexe": "Genre"},
             color_discrete_sequence=px.colors.qualitative.Pastel,
             orientation="h")            # Optionnel mais explicite pour l'horizontale

# Mise à jour du design
fig.update_layout(
    xaxis_title="Taux de Réussite (%)",
    yaxis_title="Académie",
    showlegend=True,                     # Réactivé car utile avec l'argument 'color'
    template="plotly_white",
    height=800                           # Augmenté pour éviter que les académies soient trop serrées
)

# Mise à jour du design avec une hauteur doublée
fig.update_layout(
    height=2000,                  # Hauteur augmentée (ex: 1000 pixels)
    xaxis_title="Taux de Réussite (%)",
    yaxis_title="Académie",
    showlegend=True,
    template="plotly_white"
)


fig.show()
```
